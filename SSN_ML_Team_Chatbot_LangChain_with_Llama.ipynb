This code defines a function get_prompt and necessary constants for generating prompts for model input. The get_prompt function combines an instruction with a system prompt using specified delimiters and returns the resulting prompt template.

Instruction prompt - Defines Task                 
System Prompt - Defines Rules To Govern It




```python
import json
import textwrap

B_INST, E_INST = "[INST]", "[/INST]"
B_SYS, E_SYS = "<<SYS>>\n", "\n<</SYS>>\n\n"
DEFAULT_SYSTEM_PROMPT = """\
You are a helpful, respectful and honest assistant. Always answer as helpfully as possible, while being safe. Your answers should not include any harmful, unethical, racist, sexist, toxic, dangerous, or illegal content. Please ensure that your responses are socially unbiased and positive in nature.

If a question does not make any sense, or is not factually coherent, explain why instead of answering something not correct. If you don't know the answer to a question, please don't share false information."""



def get_prompt(instruction, new_system_prompt=DEFAULT_SYSTEM_PROMPT ):
    SYSTEM_PROMPT = B_SYS + new_system_prompt + E_SYS
    prompt_template =  B_INST + SYSTEM_PROMPT + instruction + E_INST
    return prompt_template

```

## Setting up with LangChain

HuggingFacePipeline: This module provides a pipeline interface for interacting with Hugging Face models.

```python
from langchain import HuggingFacePipeline
from langchain import PromptTemplate,  LLMChain
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory, ConversationBufferWindowMemory
```

```python
template = get_prompt(instruction, system_prompt)
print(template)

prompt = PromptTemplate(template=template, input_variables=["context", "question"])
```


Create a window-based conversation memory object named memory. The parameters are configured as follows:

memory_key: Specifies the key to access the conversation history within the memory object.   

k: Determines the size of the window, specifying the maximum number of messages to retain in the conversation history.       
                
return_messages: Indicates whether to return the individual messages stored in the memory or just the entire conversation history.

```python
memory = ConversationBufferWindowMemory(
    memory_key="chat_history", k=5,
    return_messages=True
)
```


Convert the Chroma vector store object db into a retriever object named retriever. This retriever can be used to retrieve relevant documents or text segments based on vector similarity.

```python
retriever = db.as_retriever()
```


Define a function named create_pipeline that creates a text generation pipeline. The parameters are configured as follows:

max_new_tokens: Specifies the maximum number of tokens the model can generate in response to input prompts.

model: Represents the pre-trained language model to be used in the pipeline.

tokenizer: Represents the tokenizer associated with the language model.

temperature: Specifies the "temperature" parameter for controlling the randomness of generated text.

The function returns the configured pipeline object for text generation.

```python
def create_pipeline(max_new_tokens=512):
    pipe = pipeline("text-generation",
                model=model,
                tokenizer = tokenizer,
                max_new_tokens = max_new_tokens,
                temperature = 0.7)
    return pipe
```

Define a class named MedBot with the following attributes and methods:

__init__: Initializes the MedBot object with the specified parameters:

memory: Represents the conversation memory buffer.

prompt: Represents the prompt template for generating prompts in the chatbot.

task: Specifies the task type for the chatbot, defaulted to "text-generation".

retriever: Represents the document retriever for retrieving relevant documents or text segments.

create_chat_bot: Defines a method to create a chatbot instance. It configures a pipeline for text generation using the create_pipeline function, creates a Hugging Face pipeline, initializes a conversational retrieval chain from the Hugging Face pipeline, and returns the configured chatbot instance.

```python
class MedBot:
  def __init__(self, memory, prompt, task:str = "text-generation", retriever = retriever):
    self.memory = memory
    self.prompt = prompt
    self.retriever = retriever



  def create_chat_bot(self, max_new_tokens = 512):
    hf_pipe = create_pipeline(max_new_tokens)
    llm = HuggingFacePipeline(pipeline =hf_pipe)
    qa = ConversationalRetrievalChain.from_llm(
      llm=llm,
      retriever=self.retriever,
      memory=self.memory,
      combine_docs_chain_kwargs={"prompt": self.prompt}
  )
    return qa
```

Create an instance of the MedBot class named med_bot with the specified parameters:

memory: The conversation memory buffer.        
prompt: The prompt template for generating prompts in the chatbot.

```python
med_bot = MedBot(memory = memory, prompt = prompt)
```

Create a chatbot instance bot using the create_chat_bot method of the med_bot object. This method configures a chatbot with the specified conversation memory, prompt template, and other necessary components.

```python
bot = med_bot.create_chat_bot()
```


Import the gradio library for creating interactive web interfaces. Also, define two utility functions:

clear_llm_memory: Clears the long-term memory of the language model used in the chatbot.
update_prompt: Updates the prompt used by the chatbot for generating responses based on the system prompt provided.

```python
import gradio as gr
import random
import time

def clear_llm_memory():
  bot.memory.clear()

def update_prompt(sys_prompt):
  if sys_prompt == "":
    sys_prompt = system_prompt
  template = get_prompt(instruction, sys_prompt)

  prompt = PromptTemplate(template=template, input_variables=["context", "question"])

  bot.combine_docs_chain.llm_chain.prompt = prompt
```

```python
with gr.Blocks() as demo:
    update_sys_prompt = gr.Textbox(label = "Update System Prompt")
    chatbot = gr.Chatbot(label="Medical Bot", height = 300)
    msg = gr.Textbox(label = "Question")
    clear = gr.ClearButton([msg, chatbot])
    clear_memory = gr.Button(value = "Clear LLM Memory")


    def respond(message, chat_history):
        bot_message = bot({"question": message})['answer']
        chat_history.append((message, bot_message))
        return "", chat_history

    msg.submit(respond, inputs=[msg, chatbot], outputs=[msg, chatbot])
    clear_memory.click(clear_llm_memory)
    update_sys_prompt.submit(update_prompt, inputs=update_sys_prompt)

demo.launch(share=True, debug=True)
```
